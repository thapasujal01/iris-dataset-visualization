```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv"
df = pd.read_csv(url)
df.head()

# Calculate the maximum price for each neighbourhood group
max_price = df.groupby('neighbourhood_group')['price'].max()

plt.figure()
max_price.plot(kind="bar")
plt.title("Maximum Price by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Price ($)")

for index, value in enumerate(max_price):
    plt.text(index, value, f"${value}", ha='center', va='bottom')

plt.show()
avg_price = df.groupby(
    ["neighbourhood_group", "room_type"]
)["price"].mean().unstack()

plt.figure()
avg_price.plot(kind="bar")
plt.title("Average Price by Neighbourhood Group and Room Type")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price ($)")
plt.legend(title="Room Type")

plt.show()
```
