```python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()

project_id = 'bike-shareanalysis'
client = bigquery.Client(project=project_id)
table_id = "bigquery-public-data.austin_bikeshare.bikeshare_trips"
query_count = f"SELECT COUNT(*) as total_entries FROM `{table_id}`"
df_count = client.query(query_count).to_dataframe()

print(f"Total entries: {df_count['total_entries'][0]}")
query_max = f"SELECT MAX(duration_minutes) as max_duration FROM `{table_id}`"
df_max = client.query(query_max).to_dataframe()

print(f"Longest trip: {df_max['max_duration'][0]} minutes")
query_avg = f"SELECT AVG(duration_minutes) as avg_duration FROM `{table_id}`"
df_avg = client.query(query_avg).to_dataframe()

print(f"Average trip time: {df_avg['avg_duration'][0]:.2f} minutes")
```
