```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv"

df = pd.read_csv(url)

df

```

```python
plt.figure(figsize=(9,7))

for name, group in df.groupby("Class"):
    plt.scatter(group["SepalLength"], group["SepalWidth"], label=name, alpha=0.8)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris: Scatterplot Sepal Length vs Width ")
plt.legend(title="Class")
plt.show()
```
