```python
import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films"

# 1. Create a header to mimic a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
}


response = requests.get(url, headers=headers)


tables = pd.read_html(response.text, match="Phase One")
mcu_phase_1 = tables[0]


mcu_phase_1.columns = [col[-1] if isinstance(col, tuple) else col for col in mcu_phase_1.columns]


mcu_phase_1.to_csv("mcu_phase_1.csv", index=False)

print("Success! The Bifrost is open. Check your folder for 'mcu_phase_1.csv'.")
mcu_phase_1.head()
```
