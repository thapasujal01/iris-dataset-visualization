```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv"
iris = pd.read_csv(url)


iris.head()
```

```python
iris.info()
iris.describe()
```

```python
species_counts = iris['Class'].value_counts()
plt.figure(figsize=(6,4))
species_counts.plot(kind='bar')
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Count of Iris Class")
plt.show()
```

```python
iris.columns
```

```python
plt.figure(figsize=(6,4))
plt.hist(iris['SepalLength'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(iris['SepalLength'])
plt.ylabel("Sepal Length (cm)")
plt.title("Box Plot of Sepal Length")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.violinplot(iris['SepalLength'])
plt.ylabel("Sepal Length (cm)")
plt.title("Violin Plot of Sepal Length")
plt.show()
```
