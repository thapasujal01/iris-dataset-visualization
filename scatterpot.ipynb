```python
import pandas as pd
import plotly.express as px

# Load the data
url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/refs/heads/main/Data_Sets/Activity_Dataset_V1.csv"
df = pd.read_csv(url)

# 1. Data Cleaning: Drop the index column if present
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# 2. Variable Selection: Maximizing complexity for the "Bonus"
# We are using 7 variables in total here.
fig = px.scatter_matrix(
    df,
    dimensions=['distance', 'calories', 'avg_heart_rate', 'max_heart_rate', 'time'],
    color='workout_type',  # 6th Variable
    # symbol='gender',       # 7th Variable - Removed as 'gender' column does not exist
    title='Comprehensive Association Matrix of Exercise Metrics',
    labels={col: col.replace('_', ' ').title() for col in df.columns},
    opacity=0.7,
    width=1100,
    height=900
)

# 3. Refinement: Clean up the diagonal and markers
fig.update_traces(diagonal_visible=False, marker=dict(size=5))
fig.update_layout(font_size=10)

fig.show()
```
